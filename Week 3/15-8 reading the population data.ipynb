## Reading mukim sheet

```python
# import pandas as pd

# Specify the path to the Excel file
excel_file_path = "population.xlsx"

# Specify the sheet name you want to read
sheet_name = "mukims"

# Read the Excel file and the specified sheet into a DataFrame
df = pd.read_excel(excel_file_path, sheet_name=sheet_name)  # Note the 'header' parameter

# Now you have your data in the 'df' DataFrame
# print(df)
df.head()
```

```python
# print(df)
df.info()
```

```python
# Group by 'state' and calculate the sum of 'reg_total'
state_total = df.groupby('state')['reg_total'].sum()

print(state_total)
```

```python
# Group by 'state' and calculate the sum of 'reg_total'
state_total = df.groupby('district')['reg_total'].sum()

print(state_total)
```
