```python
import pandas as pd

# Specify the path to the CSV file
csv_file_path = 'FET_final_16_8.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Now you have your data in the 'df' DataFrame
# print(df)
df.info()
```

```python
# import pandas as pd

# Assuming you have a DataFrame named 'df'
# Add a new column 'tertiary address' filled with empty strings
df['tertiary address'] = ''

# Print the DataFrame to see the changes
# print(df)
df.info()
```

```python
# import pandas as pd

# Assuming you have a DataFrame named 'df'
# and you want to search for a keyword in the 'Company name' column
keyword = 'cohu'

# Filter the DataFrame to get rows where 'Company name' contains the keyword
matching_companies = df[df['Company name'].str.contains(keyword, case=False)]

# List the company names that match the keyword
matched_company_names = matching_companies['Company name'].tolist()
print(matched_company_names)
```

```python
# import pandas as pd

# Assuming you have a DataFrame named 'df'
# and you want to update the 'tertiary address' column for matched company names
# matched_company_names = ['Company A', 'Company B', 'Company C']  # Example list of matched company names
new_tertiary_address = '9Th Floor Menara Sarawak Energy No.1 The Isthmus 93050 Kuching Sarawak'

# Loop through matched company names and update the 'tertiary address' column
for company_name in matched_company_names:
    df.loc[df['Company name'] == company_name, 'tertiary address'] = new_tertiary_address

# Print the updated DataFrame
# print(df)

    # Filter rows based on the specified company name
    target_company_rows = df[df['Company name'] == company_name]

    # Set the display option to show all the content of the columns
    pd.set_option('display.max_colwidth', None)

    # Print the entire rows for the target company
    print(target_company_rows)
```

```python
# specific_df_drop_2
# Save the updated DataFrame to a CSV file
output_csv_path = 'FCM_JOBS_PAIRED_with_tertiary_address.csv'
df.to_csv(output_csv_path, index=False)

print(f"New DataFrame saved to '{output_csv_path}'")
```
