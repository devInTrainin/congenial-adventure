```python
import os
import pandas as pd
from fuzzywuzzy import fuzz

# Replace 'folder_path' with the actual path to your 'ITP companies' folder
folder_path = 'ITP companies'

# Replace 'file_name' with the actual name of the Excel file
file_name = 'FCI ITP.xlsx'

# Combine folder path and file name to get the complete file path
file_path = os.path.join(folder_path, file_name)

# Function to check if two strings are similar using fuzzywuzzy
def is_similar(str1, str2, threshold=95):
    return fuzz.partial_ratio(str1.lower(), str2.lower()) >= threshold

# Check if the file exists
if os.path.exists(file_path):
    # Read the Excel file into a pandas DataFrame
    df = pd.read_excel(file_path)

    # Deduplicate similar company names
    similar_companies = {}
    for i, company_name in enumerate(df['Company Name']):
        for name, idx in similar_companies.items():
            if is_similar(company_name, name):
                df.loc[i, 'Company Name'] = name
                break
        else:
            similar_companies[company_name] = i

    # Print the total number of unique companies after deduplication
    unique_companies = df['Company Name'].nunique()
    print('Unique Companies (after deduplication):', unique_companies)

    # Find non-unique companies
    non_unique_companies = df['Company Name'].value_counts()[df['Company Name'].value_counts() > 1]

    # Print the non-unique companies
    print('Non-Unique Companies:')
    print(non_unique_companies)
else:
    print(f"File '{file_name}' does not exist in the specified folder.")
```

```python
import os
import pandas as pd
import recordlinkage

# Replace 'folder_path' with the actual path to your 'ITP companies' folder
folder_path = 'ITP companies'

# Replace 'file_name' with the actual name of the Excel file
file_name = 'FCI ITP.xlsx'

# Combine folder path and file name to get the complete file path
file_path = os.path.join(folder_path, file_name)

# Check if the file exists
if os.path.exists(file_path):
    # Read the Excel file into a pandas DataFrame
    df = pd.read_excel(file_path)

    # Create an indexer for record linkage
    indexer = recordlinkage.Index()
    candidate_links = indexer.full()

    # Create a comparison object
    compare_cl = recordlinkage.Compare()

    # Add comparison functions for the company name column
    compare_cl.string('Company Name', 'Company Name', method='jarowinkler', threshold=0.85)

    # Compute the potential matches
    potential_matches = compare_cl.compute(candidate_links, df)

    # Use the potential matches to create a deduplication key
    dedup_key = potential_matches.sum(axis=1) > 0

    # Deduplicate the DataFrame based on the deduplication key
    deduplicated_df = df.loc[dedup_key]

    # Print the total number of unique companies after deduplication
    unique_companies = deduplicated_df['Company Name'].nunique()
    print('Unique Companies (after deduplication):', unique_companies)

    # Find non-unique companies
    non_unique_companies = deduplicated_df['Company Name'].value_counts()[deduplicated_df['Company Name'].value_counts() > 1]

    # Print the non-unique companies
    print('Non-Unique Companies:')
    print(non_unique_companies)
else:
    print(f"File '{file_name}' does not exist in the specified folder.")
```
