```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Not Enough Information', 'Information Found', 'International Company', 'Critical Issues','Partial Information']
sizes = [15, 716, 46, 2, 6]  
colors = ['skyblue', 'floralwhite', 'khaki', 'firebrick','grey']  # Colors for each slice

# Create a figure and axis with custom size
fig, ax = plt.subplots(figsize=(8, 8))

# Create a pie chart with custom colors
wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')

# Customizing font size of the labels and percentage values
for text in texts + autotexts:
    text.set_fontsize(12)

# Add a title
plt.title('Company Addresses')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
# labels = ['Not Enough Information', 'Information Found', 'International Company', 'Critical Issues', 'Partial Information']
# sizes = [15, 716, 46, 2, 6]
# colors = ['skyblue', 'floralwhite', 'khaki', 'firebrick', 'grey']  # Colors for each slice

labels = ['Information Found', 'International Company','Not Enough Information', 'Partial Information','Critical Issues']
sizes = [716, 46, 15, 6, 2]
colors = ['floralwhite','khaki','skyblue','grey','firebrick'] 

# Calculate the total sum of sizes
total_size = sum(sizes)

# Calculate the percentages relative to the total size
percentages = [size / total_size * 100 for size in sizes]

# Create a figure and axis with custom size
fig, ax = plt.subplots(figsize=(8, 8))

# Create a pie chart with custom colors
wedges, _ = ax.pie(sizes, labels=None, startangle=140, colors=colors)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a title
plt.title('Company Addresses')

# Create a legend in the top right corner
legend = ax.legend(wedges, labels, loc='upper right', bbox_to_anchor=(1, 0, 0.5, 1), title="Categories")

# Add percentage labels to the legend
for text, percentage in zip(legend.get_texts(), percentages):
    text.set_text(f'{text.get_text()} ({percentage:.1f}%)')

# Show the plot
plt.show()
```
