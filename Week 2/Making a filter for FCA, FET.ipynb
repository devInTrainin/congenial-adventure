```python
# new_df_saved = new_df

new_df_add_columns = new_df

# Regular expression patterns for email and phone number
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'
# phone_pattern = r'\b\d{3}-\d{7,8}\b'
# phone_pattern = r'\b(?:\+\d{1,2})?[-.\s]?\d{3,5}[-.\s]?\d{6,8}\b'
phone_pattern = r'(?:(?:\+\d{1,2}[-.\s]?)?(?:\d{3,5}[-.\s]?\d{6,8}|\(\d{3,5}\)[-\s]?\d{6,8}))'


# Initialize new columns
new_df_add_columns['email'] = ''
new_df_add_columns['Phone number'] = ''

# Iterate through rows and extract email and phone number
for index, row in new_df_add_columns.iterrows():
    contact_info = row['Contact Info']
    
    email_match = re.search(email_pattern, contact_info)
    phone_match = re.search(phone_pattern, contact_info)
    
    if email_match:
        new_df_add_columns.at[index, 'email'] = email_match.group()
    if phone_match:
        new_df_add_columns.at[index, 'Phone number'] = phone_match.group()

# Drop the original 'Contact Info' column
# new_df.drop(columns=['Contact Info'], inplace=True)

# Display the modified DataFrame
# print(new_df_add_columns)
new_df_add_columns
```

```python
# import pandas as pd

# Create a new DataFrame for sorting and handling duplicates
sorted_df = new_df_add_columns.copy()

# Sort the DataFrame by 'Org Name'
sorted_df.sort_values(by='Org Name', inplace=True)

# Initialize a dictionary to store unique organizations with their email and phone information
unique_orgs = {}

# Iterate through rows and update the dictionary
for index, row in sorted_df.iterrows():
    org_name = row['Org Name']
    email = row['email']
    phone = row['Phone number']
    
    if org_name in unique_orgs:
        # Append email and phone information if they're different
        if email and email not in unique_orgs[org_name]['email']:
            unique_orgs[org_name]['email'] += ', ' + email
        if phone and phone not in unique_orgs[org_name]['Phone number']:
            unique_orgs[org_name]['Phone number'] += ', ' + phone
    else:
        # Initialize entry in the dictionary
        unique_orgs[org_name] = {'Org Name': org_name, 'email': email, 'Phone number': phone}

# Create a new DataFrame from the dictionary
unique_orgs_df = pd.DataFrame.from_dict(unique_orgs, orient='index')

# Display the resulting DataFrame
print(unique_orgs_df)

```

```python
# import pandas as pd

# Assuming you have a DataFrame named 'df'

def combine_contacts_and_drop_duplicates(dataframe, company_name):
    # Find rows with the specified company name
    company_rows = dataframe[dataframe['Company'] == company_name]
    
    if len(company_rows) > 1:
        # Combine 'Contact Person' and 'Company Contact' with semicolon separator
        combined_contact_person = '; '.join(company_rows['Contact Person'])
        combined_company_contact = '; '.join(company_rows['Company Contact'])
        
        # Create a new row with combined values
        new_row = {
            'Company': company_name,
            'Contact Person': combined_contact_person,
            'Company Contact': combined_company_contact
        }
        
        # Drop original rows with the specified company name
        dataframe = dataframe[dataframe['Company'] != company_name]
        
        # Append the new combined row
        dataframe = dataframe.append(new_row, ignore_index=True)
    return dataframe

# Example DataFrame
# data = {
#     'Company': ['Company A', 'Company B', 'Company A', 'Company C'],
#     'Contact Person': ['Person 1', 'Person 2', 'Person 3', 'Person 4'],
#     'Company Contact': ['Contact 1', 'Contact 2', 'Contact 3', 'Contact 4']
# }
# df = pd.DataFrame(data)

# Input the company name to process
# Keysight Technologies Malaysia Sdn Bhd
# target_company = input("Enter the company name: ")
target_company = input("Enter the company name: ")

# Apply the function to combine contacts and drop duplicates
df = combine_contacts_and_drop_duplicates(df, target_company)

# Display the modified DataFrame
print(df)

#enter Keysight Technologies Malaysia Sdn Bhd
```
