```python
import pandas as pd
import folium
```

```python
data = pd.read_excel("negeri split_no filters.xlsx", sheet_name='Alyaa')
```

```python
data.head()
```

```python
data.info()
```

```python
target_column = 'STATE'
```

```python
target_word = 'labuan'

data_labuan = data[data[target_column].str.contains(target_word, case=False, na=False)]
data_labuan.head()
```

```python
target_word = 'johor'

data_johor = data[data[target_column].str.contains(target_word, case=False, na=False)]
data_johor
```

```python
target_word = 'terengganu'

data_terengganu = data[data[target_column].str.contains(target_word, case=False, na=False)]
data_terengganu
```

```python
target_word = 'melaka'

data_melaka = data[data[target_column].str.contains(target_word, case=False, na=False)]
data_melaka
```

```python
#latitude n longitude for Labuan 5.3211936, 115.2219741
labuan_latitude = 5.3211936
labuan_longitude = 115.2219741

labuan_map = folium.Map(location=[labuan_latitude, labuan_longitude], zoom_start=12)

for index, row in data_labuan.iterrows():
    popup_content = f"<strong>{row['Company name']}</strong><br>{row['Company address']}"
    
    folium.Marker(
        location = [row['map_latitude'], row['map_longitude']],
        popup = folium.Popup(popup_content),
        tooltip = row['Company name']
    ).add_to(labuan_map)
    
labuan_map
```
